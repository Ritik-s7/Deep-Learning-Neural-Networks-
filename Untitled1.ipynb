```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observation = 1000
```

```python
xs = np.random.uniform(-10,10,(observation,1))
zs = np.random.uniform(-10,10,(observation,1))
```

```python
print(xs.shape, zs.shape)
```

```python
x = np.column_stack((xs,zs))
```

```python
x.shape 
```

```python
noise = np.random.uniform(-1,1,(observation,1))
targets = 2*xs + 5*zs - 6 + noise
```

```python
targets.shape
```

```python
weights = np.random.uniform(-1,1,(2,1))
biases = np.random.uniform(-1,1,(1,1))
```

```python
print(weights)
print(biases)
```

```python
learning_rate = 0.01
```

```python
for i in range(observation):
    output = np.dot(x,weights) + biases
    deltas = output-targets
    loss = np.sum(deltas**2)/2/observation
    print(loss)
    
    deltas_scaled = deltas/observation
    
    weights = weights - learning_rate*(np.dot(x.T,deltas_scaled))
    biases = biases - learning_rate * np.sum(deltas_scaled)
    
```

```python
print(weights)
print(biases)
```
